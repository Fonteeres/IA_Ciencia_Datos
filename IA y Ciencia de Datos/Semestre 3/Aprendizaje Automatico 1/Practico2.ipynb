```python
import pandas as pd 
import numpy as np
df_vino = pd.read_csv('/content/drive/MyDrive/Aprendizaje_Automatico_1/set_datos/winequality-white.csv', sep=';')
df_vino
```

#Separo X e Y con Pandas

```python
df_vino_x = df_vino.drop(["quality"], axis = 1)
df_vino_x
```

```python
df_vino_y = df_vino['quality']
df_vino_y
```

# Separo En X_train, X_test, y_train, y_test y el validation

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_vino_x, df_vino_y, test_size=0.2, random_state=123)
#subdivido el X_train e y_train en los Validations
Valx_train, Valx_test, Valy_train, Valy_test = train_test_split(X_train, y_train, test_size=0.2, random_state=123)
```

#Clasificacion


## Arbol de Clasificacion

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
rango = np.arange(5,18,1)#una variable para medir la profundidad
splitterA = ['best','random']#Con esta mido el tipo de division
acmax = 0 # un contador para saber cual es el mejor Acuracy
#un For para pobrar los mejores valores del modelo
for x in rango:
  for j in splitterA:
        ArbolClasificacion = DecisionTreeClassifier(max_depth = x , splitter = j )
        ArbolClasificacion.fit(X_train,y_train)
        ArbolPrediccion =  ArbolClasificacion.predict(Valx_test)
        ac = accuracy_score(ArbolPrediccion,Valy_test)
        score = ArbolClasificacion.score(X_test, y_test)
        print('El porcentaje de Acurracy es de: ', ac , " La profundida del arbol es de:", x, " El splitter o divisor es de tipo:", j, ' Y su Score es de:',score)
        if ac > acmax:
          acmax = ac
          Xfinal = x
          Jfinal = j
          ScoreF = score
print('El porcentaje de Acurracy es de: ', acmax , " La profundida del arbol es de:", Xfinal, " El splitter o divisor es de tipo:", Jfinal, ' Y su Score es de:',ScoreF)
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
rango = np.arange(10,25,1)#una variable para medir la cantidad de vecinos
weightsA = ['uniform','distance']#de termino si los datos tienen peso en base a su distancia o no
acmax = 0 # un contador para saber cual es el mejor Acuracy
#un For para pobrar los mejores valores del modelo
for x in rango:
  for j in weightsA:
        KNNClasificacion =  KNeighborsClassifier(n_neighbors= x , weights= j )
        KNNClasificacion.fit(Valx_train,Valy_train)
        KNNPrediccion =  KNNClasificacion.predict(Valx_test)
        ac = accuracy_score(KNNPrediccion,Valy_test)
        score = KNNClasificacion.score(X_test, y_test)
        print('El porcentaje de Acurracy es de: ', ac , " La cantidad de vecinos es de:", x, " El peso de la distancia es de tipo:", j, ' Y su Score es de:',score)
        if ac > acmax:
          acmax = ac
          Xfinal = x
          Jfinal = j
          ScoreF = score
print('El porcentaje de Acurracy es de: ', acmax , " La cantidad de vecinos es de:", Xfinal, " El peso de la distancia es de tipo:", Jfinal, ' Y su Score es de:',ScoreF)
#claramente sufro de overfitting pero no se come medir en que momento exacto llego al momento de equilibrio
```

## SVM Regresion




```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
regularizacion = np.arange(1.0,5.0)#una variable para medir la regularizacion
KernelA = ['linear','poly','rbf','sigmoid']#Con esta mido el tipo de division
acmax = 0 # un contador para saber cual es el mejor Acuracy
#un For para pobrar los mejores valores del modelo
for x in regularizacion:
  for j in KernelA:
      SVC = SVC(C=x, kernel=j, gamma='auto')
      SVC.fit(Valx_train,Valy_train)
      SVCPrediccion = SVC.predict(Valx_test)
      ac = accuracy_score(SVCPrediccion,Valy_test)
      score = SVC.score(Valx_test,Valy_test)
      print('El porcentaje de Acurracy es de: ', ac , " La profundida del arbol es de:", x, " El splitter o divisor es de tipo:", j, ' Y su Score es de:',score)
      if ac > acmax:
          acmax = ac
          Xfinal = x
          Jfinal = j
          ScoreF = score
print('El porcentaje de Acurracy es de: ', acmax , " La profundida del arbol es de:", Xfinal, " El splitter o divisor es de tipo:", Jfinal, ' Y su Score es de:',ScoreF)
```
