### Ejercicio 17

La columna **budget** corresponde al **Presupuesto** y la columna **gross** corresponde a los **ingresos**, queremos agregar una columna al DataFrame donde calculemos las **ganancias** (que se calculara como ingresos menos presupuesto). Llmela ganancias y agrguela a df_movies.

Para finalizar guarde las columnas movie_title, director_name, budget, gross y ganancias como df_money y muestre los 5 primeros elementos 

```python
# Llamo al Df le creo la nueva culmna y le asgino la resta entre el ingreso menos el presupuesto
df_movies['ganancias'] = df_movies.gross - df_movies.budget
#muestro el df
df_movies
```

```python
# aca creo el Df y le asigno las columnas
df_money=pd.DataFrame(df_movies[['movie_title','director_name','budget','gross','ganancias']])
#muestro el data frame
df_money.head(5)
```

### Ejercicio 18
A partir de df_money muestre cules son las 10 pelculas con mayores ganancias? Gurdelo en df_topten_movies, grabe los datos en un archivo dentro de la carpeta "datos" con el nombre "Ejericio18.csv" y muestre los resultados.

```python
# Creo el Df top ten y le asigno las 10 peliculas mas taquilleras
df_topten_movies = pd.DataFrame(df_money.sort_values('ganancias',ascending = False).head(10))
#lo guardo en un archivo csv
df_topten_movies.to_csv('datos/Ejercicio18.csv')
#lo muestro
df_topten_movies
```

### Ejercicio 20
A partir del DataFrame df_topten_movies del ejercicio 18:
* Realice un grfico de lneas con los datos del DataFrame.
* Defina la columna movie_title como ndice del DataFrame.
* El grfico debe tener el ttulo: "Pelculas con mayor ganancia".
* Configure la leyenda para que se visualice en la parte superior derecha.
* Asignele al grfico una rejilla o grilla para las coordenadas x e y.
* Guarde el grfico, en la carpeta "datos", en un archivo: "Ejercicio20.png"
