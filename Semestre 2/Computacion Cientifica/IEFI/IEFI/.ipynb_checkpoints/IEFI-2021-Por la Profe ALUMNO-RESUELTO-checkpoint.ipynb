### Ejercicio 17

La columna **budget** corresponde al **Presupuesto** y la columna **gross** corresponde a los **ingresos**, queremos agregar una columna al DataFrame donde calculemos las **ganancias** (que se calcularía como ingresos menos presupuesto). Llámela ganancias y agréguela a df_movies.

Para finalizar guarde las columnas movie_title, director_name, budget, gross y ganancias como df_money y muestre los 5 primeros elementos 

```python
# Escriba aquí su código
import pandas as pd
df_movies =pd.read_csv("datos/movie.csv",sep=',')

df_movies['ganancias'] = df_movies.gross - df_movies.budget

df_money = df_movies[['movie_title', 'director_name', 'budget', 'gross','ganancias']]
df_money.head()
```

### Ejercicio 18
A partir de df_money muestre cuáles son las 10 películas con mayores ganancias? Guárdelo en df_topten_movies, grabe los datos en un archivo dentro de la carpeta "datos" con el nombre "Ejercicio18.csv" y muestre los resultados.

```python
# Escriba aquí su código
import pandas as pd
df_movies =pd.read_csv("IEFI/datos/movie.csv",sep=',')
df_movies['ganancias'] = df_movies.gross - df_movies.budget
df_money = df_movies[['movie_title', 'director_name', 'budget', 'gross','ganancias']]

df_topten_movies = df_money.sort_values(by=['ganancias'],ascending=[False]).head(10)
df_topten_movies.to_csv('Ejercicio18.csv',sep=',')
df_topten_movies
```

### Ejercicio 19

El directivo de una compañía cinematográfica quiere saber cuáles son las ganancias promedio que puede esperar obtener con una nueva película. Guarde el resultado en la variable prom_ganancias y muestre el resultado 

```python
# Escriba aquí su código
import pandas as pd
df_movies =pd.read_csv("datos/movie.csv",sep=',')
df_movies['ganancias'] = df_movies.gross - df_movies.budget
df_money = df_movies[['movie_title', 'director_name', 'budget', 'gross','ganancias']]

prom_ganancias = df_money.ganancias.mean()
prom_ganancias
```

### Ejercicio 20
A partir del DataFrame df_topten_movies del ejercicio 18:
* Realice un gráfico de líneas con los datos del DataFrame.
* Defina la columna movie_title como índice del DataFrame.
* El gráfico debe tener el título: "Películas con mayor ganancia".
* Configure la leyenda para que se visualice en la parte superior derecha.
* Asignele al gráfico una rejilla o grilla para las coordenadas x e y.
* Guarde el gráfico, en la carpeta "datos", en un archivo: "Ejercicio20.png"
* Muestre el gráfico

```python
# Escriba aquí su código
import matplotlib.pyplot as mp
import pandas as pd
df_movies =pd.read_csv("datos/movie.csv",sep=',')
df_movies['ganancias'] = df_movies.gross - df_movies.budget

df_money = df_movies[['movie_title','ganancias']]

df_topten_movies = df_money.sort_values(by=['ganancias'],ascending=[False]).head(10)

df=df_topten_movies.set_index('movie_title')
fig, eje = mp.subplots()
df.plot(ax = eje)
# Título del gráfico
eje.set_title('Películas con mayor ganancia')
# Ubicación de la leyenda
eje.legend(loc = 'upper right')
#Rejilla
eje.grid(axis = 'both', color = 'gray', linestyle = 'dashed')
mp.show()


```
