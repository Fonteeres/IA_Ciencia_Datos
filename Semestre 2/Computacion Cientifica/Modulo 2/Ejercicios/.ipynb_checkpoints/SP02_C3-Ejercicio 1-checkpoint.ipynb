```python
# Abra el archivo C1-Empleados.csv
import pandas as pd
df_empleados = pd.read_csv('Datos/C1-Empleados.csv', sep=';')
df_empleados.head()
```

```python
# Calcule cuanto pagar de antiguedad si es 
# el 1,5 del básico por cada año de antiguedad
Valor = (df_empleados.Antigüedad * 1.5) * df_empleados.Básico
```

```python
#Asigne ese resultado al df
df_empleados['AntiguedadSobreBásico'] = Valor
df_empleados
```

```python
#Calcular descuento aporte
#es el 18% del básico
Descuento = df_empleados.Básico * 0.18
```

```python
#Asignarlo a la columna aporte
df_empleados['Aporte'] = Descuento
df_empleados
```

```python
#Calcular total a cobrar: básico+antiguedad-aportes
Total = df_empleados.Básico + df_empleados.AntiguedadSobreBásico - df_empleados.Aporte
```

```python
#Asignarlos
df_empleados['Total_a_Cobrar'] = Total
```

```python
df_empleados
```

```python
#Mostrar empleados de la categoría 1
df_empleados[df_empleados.Categoría <= 1]
```
