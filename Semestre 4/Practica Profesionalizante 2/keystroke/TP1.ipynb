```python
df_Tiempo_Tecla = df_Tiempo_Tecla.drop('tiempo_id',axis=1)
```

```python
df_final = pd.DataFrame(columns=['tiempo_entre_tecla_up','tiempo_entre_tecla_down'])
```

```python
df_final['tiempo_entre_tecla_down']  = df_Tiempo_Tecla.groupby(['usuario','practica','modo']).apply(lambda x: x['down'] - x['anterior_down'])
df_final['tiempo_entre_tecla_up']  = df_Tiempo_Tecla.groupby(['usuario','practica','modo']).apply(lambda x: x['up'] - x['anterior_down'])
```
