```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
df_Cambio_mano = pd.read_csv('Datos\cidia_pp2_table_tiempos_cambio_mano.csv')
df_Tiempo_Tecla = pd.read_csv('Datos\cidia_pp2_table_tiempos_por_tecla.csv')
df_Usuario = pd.read_csv('Datos\cidia_pp2_table_usuarios.csv')
```

```python
df_Cambio_mano = df_Cambio_mano.drop('tiempo_id',axis=1)
```

```python
df_Cambio_mano = df_Cambio_mano[df_Cambio_mano['tiempo']<= 3000]
df_Cambio_mano = df_Cambio_mano[df_Cambio_mano['modo'] == 1]
df_Cambio_mano
```

```python
px.violin(df_Cambio_mano, y='tiempo', color='usuario')
```

```python
df_Tiempo_Tecla = df_Tiempo_Tecla.drop('tiempo_id',axis=1)
```

```python
df_Tiempo_Tecla= df_Tiempo_Tecla[df_Tiempo_Tecla['modo'] == 1]
```

```python
df_Tiempo_Tecla
```

```python
df_Tiempo_Tecla['Tiempo_Up'] = df_Tiempo_Tecla['up'] - df_Tiempo_Tecla['anterior_down']
df_Tiempo_Tecla['Tiempo_down'] = df_Tiempo_Tecla['down'] - df_Tiempo_Tecla['anterior_down']
```

```python
df_Tiempo_Tecla['Tecla_esp'] = df_Tiempo_Tecla['tecla'].map(chr)
df_Tiempo_Tecla['Tecla_esp_sig'] =  df_Tiempo_Tecla['tecla'].map(chr).shift(-1)
```

```python
df_Tiempo_Tecla['Frase_2'] = df_Tiempo_Tecla['Tecla_esp'] + df_Tiempo_Tecla['Tecla_esp_sig']
```

```python
df_Tiempo_Tecla.drop('Tecla_esp_sig',axis=1)
```

```python
df_Tiempo_Tecla_pivoteado = df_Tiempo_Tecla.pivot(columns='usuario')
df_Tiempo_Tecla_pivoteado
```

```python
df_Tiempo_Tecla.groupby(['usuario','practica','Tecla_esp'])['Tecla_esp'].count().reset_index(name='cant')
```

```python
df_Tiempo_Tecla.groupby(['usuario','practica','Frase_2'])['Frase_2'].count().reset_index(name='cant')
```
