```python
from nltk.stem.snowball import stopwords
import nltk
from nltk import sent_tokenize # separa en sentencias lo que buscamos tokenizar 'oraciones'
from nltk import regexp_tokenize # separa las sentencias en el elemento simple
from nltk import word_tokenize # 
from collections import Counter
nltk.download('punkt')
nltk.download('stopwords')
stops = set(stopwords.words('spanish'))
# buscar diferencia entre word y regexp.
```

```python
noticia= 'Chris Martin llegó a Perú junto a su banda Coldplay y causó gran emoción entre sus fanáticos, quienes contaron las horas para asistir a uno de los conciertos más esperados del año. Como se sabe, el grupo regresó a Lima para brindar dos conciertos los días 13 y 14 de septiembre. En ambos shows la peruana Andrea Martínez será la telonera de la noche y Camila Cabello la invitada especial de la banda.'
```

```python
oraciones = sent_tokenize(noticia)
```

```python
palabras = []
for oracion in oraciones:
  for palabra in regexp_tokenize(noticia, pattern='\w+'): 
      palabras.append(palabra)
len(palabras)
```

```python
Counter(palabras).most_common(10)
```

```python
filtrada = []
for palabra in palabras:
  if palabra not in stops:
    filtrada.append(palabra)
len(filtrada)
```

```python
Counter(filtrada).most_common(10)
```
