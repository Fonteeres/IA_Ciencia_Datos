# Y1_train

## SVR

```python
SVR_Model = SVR()
parametros_SVR = {'C':np.arange(0.5,5,0.5), 'degree':[3,6,9],'kernel':['poly', 'rbf','sigmoid']}
grid_svr = GridSearchCV(estimator = SVR_Model, param_grid = parametros_SVR ,scoring = 'neg_mean_squared_error', cv = 5, verbose = 1)
grid_svr.fit(X, Y1)
svr = pd.DataFrame(grid_svr.cv_results_)
```

## logistic Regression

```python
LOGR_Model = LogisticRegression()
parametros_LOGR = {'penalty':["l2"],'C':np.arange(0.1,1.1,0.1)}
grid_LOGR = GridSearchCV(estimator = LOGR_Model, param_grid = parametros_LOGR ,scoring = 'neg_mean_squared_error', cv = 5, verbose = 1)
grid_LOGR.fit(X, Y1)
LOGR = pd.DataFrame(grid_LOGR.cv_results_)
```
