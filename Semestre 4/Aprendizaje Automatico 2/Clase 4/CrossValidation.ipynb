```python
import pandas as pd
from sklearn.model_selection import cross_validate
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
df=pd.read_csv('data/datos1.csv')
df
```

```python
X = df.drop('y',axis=1)
y = df['y']
```

```python
arbol = DecisionTreeClassifier(max_depth=4)
```

```python
a = cross_val_score(arbol,X,y)
a.mean()
```

```python
a.std()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=20,random_state=123)
```

```python
arbol = DecisionTreeClassifier(max_depth=4, random_state=123)
score_Cros = cross_validate(arbol,X_train,y_train,cv=5,scoring=['accuracy','f1','precision', 'recall'])
score_Cros
```

```python
B = score_Cros['test_accuracy'].mean()
score_Cros['test_f1'].mean()
```

```python
a = 2*score_Cros['test_accuracy'].std()
```
