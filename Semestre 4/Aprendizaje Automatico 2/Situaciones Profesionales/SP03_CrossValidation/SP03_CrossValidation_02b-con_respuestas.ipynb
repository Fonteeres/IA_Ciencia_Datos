## GridSearchCV   

> GridSearchCV nos permitirá seleccionar **para un modelo dado** y usando Cross Validation el **mejor valor para su/s hiperparámetro / s** , en forma casi "automática", **nos evitará tener que hacer un for** para iterar por cada valor posible de cada hiperparámetro.  

**Nota:** Habrá notado que hemos usado el plural para hiperparámetro / s, esto es así porque puede ser que algún modelo tenga más de uno, lo cual obviamente complica su búsqueda. Hasta ahora hemos hecho ejemplos con uno solo por fines didácticos, pero a partir de ahora podremos resolver este problema fácilmente. 

Supongamos que queremos resolver el mismo problema anterior: determinar para los modelos de Árbol y Regresión Logística con Ridge,  cuáles son los mejores valores para sus respectivos hiperparámetro / s y además queremos hacerlo con Cross Validation para evaluar con más confianza los resultados, en definitiva lo que hacíamos en el tema anterior con un **for** pero ahora vamos a hacerlo con GridSearchCV.

La documentación oficial: https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html

La sintaxis es la siguiente
~~~
sklearn.model_selection.GridSearchCV(estimator, param_grid, *, scoring=None, n_jobs=None, iid='deprecated', refit=True, cv=None, verbose=0, pre_dispatch='2*n_jobs', error_score=nan, return_train_score=False)
~~~
Los parámetros principales son: 

- **estimator**: lo mismo que antes, es el modelo que queremos utilizar
- **param_grid**: Es un **diccionario** con el nombre de cada hiperparámetro y los valores que queremos que pruebe. Por ejemplo si el modelo que estamos utilizando admite 3 hiperparámetros llamados a, b y c, sería:   
~~~
{'a':[1,2,3,4], 'b':[1,10,100,1000], 'c':[0.0001,0.001,0.01]}  
~~~
donde hemos indicado para cada parámetro los valores que queremos que pruebe. **Los nombres de los hiperparámetros deben ser los que admite el modelo en cuestión** no los podemos inventar.

- **scoring:** la métrica de evaluación que deseamos utilizar, por defecto el valor es None lo cual significa que utilizará el score por defecto que tenga el modelo, que para el caso de Clasificación suele ser AC y para el de Regresión suele ser $R^2$. Pero si lo deseamos podemos indicarle otro, o incluso **varios**. 

- **refit**: Por defecto en True, nos permitirá al final calcular la métrica del modelo evaluado en todo el dataset que le hayamos pasado. Dejar en True.

- **cv:** cantidad de folds, por defecto 5 
- **return_train_score=False** False por defecto, pero puede ser de interés obtener también los valores de score correspondientes al train set.  

- **n_jobs=None** Si nuestro procesador posee más de un núcleo, podemos indicar cuántos debe usar, si el valor es -1 usa todos los que tiene.

Vamos a ver cómo lo podemos utilizar!

## Ejemplo 1 : Cross Validation para Selección de Modelo e Hiperparámetros:   "Separando el Test Set" usando GridSearchCV (más "automático").

### Cargamos los datos y separamos en Train y Test en proporción 80/20 y con semilla 123

```python
import pandas as pd
df=pd.read_csv('data/datos1.csv')


X=df.drop('y', axis='columns')
y=df['y']
```

Ahora separamos en Train y Test:

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=20,random_state=123)
```

```python
from sklearn.model_selection import GridSearchCV
```

### Modelo de Árbol:

Creamos el modelo de árbol

```python
from sklearn.tree import DecisionTreeClassifier
arbol=DecisionTreeClassifier(random_state=123)  # para que sea repetible
```

#### Hiperparámetros

**Armamos el Diccionario con los nombres y valores de los Hiperparámetros**:  

Aqui es donde podríamos pasar más de un hiperparámetro ... y nos evitaremos hacer un for dentro de otro for, etc si tenemos varios.   
En este caso sólo estamos interesados en la profuncidad, **max_depth** (este nombre debe ser el que corresponde al modelo, por las dudas, fijarse en la documentación oficial):  
https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html )


```python
parametros_arbol = {'max_depth':[1,2,3,4]}   
```

#### Creamos el GridSearchCV

```python
grid_arbol = GridSearchCV(estimator = arbol,param_grid = parametros_arbol,scoring = 'accuracy', cv = 5, verbose = 1,n_jobs = -1)

# Usé verbose=1 para que nos cuente un poco qué hace, puede omitirse.
```

#### .fit   Entrenamiento  del GridSearchCV
Lo "entrenamos" con .fit en nuestros datos.    

**Observe que le debemos pasar el Train Set para que entrene, porque queremos que cree los folds dentro de él sin tocar el Test Set que dejamos aparte. Al dividir al Train Set, cada una de las partes corresponderá a lo que llamamos ValTrain y ValTest.** que es lo que queríamos hacer.

```python
grid_arbol.fit(X_train, y_train);   # el ";" al final es para que no muestre toodos los parámetros que va a usar.
```

**Listo** Ya está todo ... **sin un for** para iterar en los valores del hiperparámetro (o varios si hubieran sido varios hiperparámetros!). Para cada valor de hiperparámetro (o combinación de ellos si eran varios) hizo los folds, entrenó en lo que llamamos ValTrain y evaluó Accuracy en el ValTest de cada fold, también determinó qué parámetro arrojó mejor resultado (Accuracy en este caso) y varias cosas más!

#### Veamos los resultados

##### .best_score_  
**Cuál fue el mejor resultado obtenido para la métrica de evaluación (en este caso Accuracy)?**

```python
AC_arbol_best=grid_arbol.best_score_
AC_arbol_best
```

##### .best_params_ 
**Cuál fue el valor de hiperparámetro/ s con el cual se obtuvo el mejor score anterior?**

**Nota:** Si hubiéramos pasado más hiperparámetros en el diccionario aqui nos hubiera dado la combinación de ellos que mejor resultado obtuvo!

```python
profundidad_arbol_best=grid_arbol.best_params_
profundidad_arbol_best
```

##### .best_estimator_

**Cuál fue el mejor modelo?**  

Aquí nos devolvuelve todo el modelo ya **entrenado y con el mejor valor de Hiperparámetro/s** en este caso profundidad 2 y el resto de los parámetros fijos que le pasamos nosotros. **Este modelo incluye todo lo necesario para usarse luego para pronosticar, etc. **

```python
modelo_arbol_best=grid_arbol.best_estimator_
modelo_arbol_best
```

##### .cv_results_  
**Queremos ver un resumen de todo lo que hizo?**

```python
grid_arbol.cv_results_
```

Observe que presenta los resultados con un diccionario, así que si queremos acceder a sólo alguno de estos valores, podemos hacerlo de la siguiente manera, seguramente estaremos interesados en dos de estos resultados, el promedio de Accuracy ('mean_test_score') y el Desvío Standard 'std_test_score', tenga en cuenta que los resultados se presentan en el orden de los valores del hiperparámetro:

```python
grid_arbol.cv_results_['params']
```

Es decir que el primer resultado corresponde a profundidad 1, el siguiente a profundidad 2, etc.

##### Accuracy promedio para cada valor del hiperparámetro:

```python
grid_arbol.cv_results_['mean_test_score']
```

Cómo se lee? 

- Para el el primer valor de hiperparámetro, es decir profundidad 1, el promedio de Accuracy en los 5 folds fue de 0.766,
- Para el segundo, profundidad 2 el valor promedio de Accuracy fue de 0.8333, y así con el resto.
Puede observar que efectivamente cuando la profundidad es 2 obtuvo una Accuracy de 0.8333 que es la mejor de todas.

##### Desvío Standard de Accuracy para cada valor del hiperparámetro:

```python
desvios_arbol=grid_arbol.cv_results_['std_test_score']
desvios_arbol
```

Se lee de la misma manera que el anterior:

Podemos ver que el desvío standard para profundidad 2 es de 0.1054

Leyendo los valores podemos escribir Promedio +/- 2Desvío Sandard  para la profundidad 2 que es la que dio mejor AC como hicimos anteriormente:

El mejor Accuracy ya lo teníamos ( en AC_arbol_best), pero el Desvío deberíamos escribirlo **manualmente** (oh, oh!):

```python
print("Mejor Accuracy: ",AC_arbol_best,' +/- ', 2*0.10540926, '( 95% )')
```

##### .best_index_
**Acceso programático al desvío correspondiente al mejor Accuracy**

Algunas personas se quejan porque dicen que hay que leer literalmente el valor del desvío para el mejor caso, sin embargo **también es posible acceder programáticamente** a cualquier valor de desvío (o de media, etc, cualquier cosa que esté en cv_results_) si uno quiere automatizar todo.   

Ocurre que GridSearchCV también tiene el atributo **best_index_**, que nos permite acceder programáticamente al valor del  desvío correspondiente al hiperparámetro que anduvo mejor!  

Fijémonos:

```python
grid_arbol.best_index_
```

Nos devolvió 1 porque si bien era el segundo elemento (profundidad 2), corresponde a índice 1 porque Python numera desde 0.  

Entonces al array  que llamamos 
~~~
desvios_arbol = grid_arbol.cv_results_['std_test_score'] 
~~~
simplemente deberíamos pasarle el índice 
~~~
[grid_arbol.best_index_] 
~~~

##### Desvío Standard de Accuracy del mejor modelo

```python
desvio_arbol_mejor=desvios_arbol[grid_arbol.best_index_]
desvio_arbol_mejor
```

##### AC +/- 2 *  desvio  (95%)

```python
print("Modelo de Árbol")
print("Mejor Accuracy: ",AC_arbol_best,' +/- ', 2*desvio_arbol_mejor, '( 95% )')
```

Y lo pudimos hacer todo programáticamente! Por supuesto no es necesario hacerlo así, tranquilamente podemos observar los resultados y tomar nota a mano. 

#### Resultados de GridSearchCV como DataFrame

Algo más arriba usamos 
~~~
grid_arbol.cv_results_
~~~

y vimos que nos presentaba todos los resultados posibles en formato de **diccionario**, si prefiere ... hasta lo puede pasar a un DataFrame!

```python
resultados_arbol=pd.DataFrame(grid_arbol.cv_results_)
resultados_arbol
```

y podemos acceder a cualquier elemento en la tradicional forma provista por Pandas!  

Por supuesto que podemos seguir usando el best_index_  en el DataFrame, como por ejemplo si queremos sólo uno de sus valores, por ejemplo el valor del hiperparámetro para el mejor caso:

```python
resultados_arbol.loc[grid_arbol.best_index_,'params']
```

O  bien podemos obtener todos los resultados correspondientes al mejor modelo!

```python
# Resúmen para el Mejor Árbol:
resultados_arbol.loc[grid_arbol.best_index_,:]
```

También puede resultar de mucho interés la última columna del DataFrame que nos ofrece un ranking para cada caso...

### Modelo de Regresión Logística + Ridge    - Autoevaluación - 


Se anima a hacer Ud solo el otro modelo que habíamos usado en el procedimiento anterior pero ahora usando GridSearchCV?  

- Era Regresión Logísitca con Regularización Ridge
- Los valores para el hiperparámetro C eran [1e-6,1e-5,1e-4,0.001,0.01,0.1,1,10]
- al GridSearchCV llámelo **grid_RL**

Obtenga lo mismo que antes:  

- Mejor valor de Accuracy +/- 2 desvíos
- Valor del Hiperparámetro C para el cual se obtuvo.
- Mejor modelo de RL + Ridge

```python
from sklearn.linear_model import LogisticRegression

# Creamos el modelo con los parámetros que no cambiarán, observe que no le pasamos el valor de C de regulrización
Log_Ridge=LogisticRegression(penalty='l2', max_iter=10000, tol=0.0001) 

# Armamos el diccionario con el  nombre y valores para los Hiperparámetros
parametros_RL = {'C':[1e-6,1e-5,1e-4,0.001,0.01,0.1,1,10]}   

# Armamos el GridSearchCV
grid_RL = GridSearchCV(estimator = Log_Ridge,param_grid = parametros_RL,scoring = 'accuracy', cv = 5, verbose = 1,
                        n_jobs = -1)

# Lo entrenamos en el Train Set (él lo dividirá internamente y en cada fold estarán en ValTrain y ValTest)
grid_RL.fit(X_train, y_train);
# Listo!
```

##### .best_score_

```python
# Obtenemos el mejor AC 
AC_RL_best=grid_RL.best_score_
AC_RL_best
```

```python
# Que se obtuvo para un valor del hiperparámetro C de:
```

##### .best_params_

```python
C_RL_best=grid_RL.best_params_
C_RL_best
```

##### .best_estimator_

**Recordemos que aquí tenemos el modelo completo con todo lo necesario inclusive para pronosticar, etc**

```python
modelo_RL_best=grid_RL.best_estimator_
modelo_RL_best
```

##### .cv_results_

##### AC +/- 2 *  desvio  (95%)

```python
print("Modelo de Regresión Logística  con Regularización Ridge")
print("Mejor Accuracy: ",AC_RL_best,' +/- ', 2*desvio_mejor_RL, '( 95% )')
```
