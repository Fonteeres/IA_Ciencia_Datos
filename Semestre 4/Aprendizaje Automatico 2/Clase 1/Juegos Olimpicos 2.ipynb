```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(8)
X = df[['Year']]
y = df[['Result']]
X = poly.fit_transform(X)
reg = LinearRegression().fit(X, y)
reg.score(X, y)
```

```python
a = reg.predict(X)
mean_squared_error(a, y, squared=False)
```

```python
from sklearn.linear_model import Ridge
clf = Ridge(alpha=5).fit(X, y)
b = clf.predict(X)
mean_squared_error(y,b,squared=False)
```

```python
from sklearn import linear_model
clfL = linear_model.Lasso(alpha=0.1).fit(X, y)
c = clf.predict(X)
mean_squared_error(y,c,squared=False)
```

```python
from sklearn.linear_model import ElasticNet
regr = ElasticNet(random_state=0,l1_ratio=0).fit(X, y)
d = regr.predict(X)
mean_squared_error(y,d,squared=False)
```
