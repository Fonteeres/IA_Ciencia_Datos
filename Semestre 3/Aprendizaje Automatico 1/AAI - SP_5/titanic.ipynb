```python
df = df.drop(['Name','Cabin'],axis=1)
df
```

```python
df.groupby('Embarked')['Embarked'].count() #descubro lo que hago abajo
```

```python
df['Embarked'] = df.groupby('Embarked')['Embarked'].transform('count')#transformo los tickets s,q,c en numericos contando la cantidad y dandole ese valor
```

```python
df = df['Embarked'].dropna()
```

```python
df['Sex'] = np.where(df['Sex']=='female',1,0)#transformo female y male a 0 y 1
df
```

```python
df.groupby('Pclass')['Age'].median()
```
