## Resolución del Ejercicio 5 sin entrenar en X_todos

Con el objetivo de evitar errores, por valores de variables que circunstancilamente no figuraran por azar en el X_train hemos procedido a entrenar en "todo" el dataset, si bien esto **no** nos traerá problemas en cuanto a "aprender" del X_train, puede sernos útil recabar previamente la información  de los valores posibles de cada variable y "pasarlos a mano". Aunque pueda parecer más lento, hay circunstancias en las cuales se justifica.  

Al pasar explícitamente los valores que pueden asumir las variables categóricas, no será necesario crear X_todos y  podremos entrenar en el X_train, ésto puede ser ventajoso en algunos casos.

> En qué orden se ubicarán las columnas después de la transformación? **En el orden en que las pasemos en la lista de valores**.

Repetiremos el ejercicio anterior, Ejercicio 5, pero ahora cambiando la opción *categories* del OneHotEncoder en vez de deajarl en 'auto', le  pasaremos una lista con los valores de las variables categóricas. 

```python
import numpy as np
import pandas as pd
from sklearn.compose import ColumnTransformer 
from sklearn.preprocessing import OneHotEncoder
```

```python
X_train5=pd.DataFrame([[1,'Lunes','0-15'],[5,'Martes','16-30'],[2,'Miércoles','16-30'],
                       [4,'Miércoles','51_o_mas'],[4,'Viernes','51_o_mas'],
                       [3,'Sábado','16-30'], [2,'Domingo','0-15'],[8,'Miércoles','51_o_mas'], 
                       [5,'Lunes','16-30'],[2,'Sábado','51_o_mas'] ], columns=['x1','dia','edad'])
X_train5
```

```python
X_test5=pd.DataFrame([[2,'Lunes','31-50'],[2,'Martes','51_o_mas'],[2,'Miércoles','0-15'],[4,'Jueves','16-30']])
X_test5.columns=['x1','dia','edad']
X_test5    
```

Ahora crearemos el transformador OneHotEncoder y le pasaremos los valores de cada una de las variables "a mano", esta es una opción válida cuando de antemano conocemos cuáles son todas las categorías que puede tomar la variable.

```python
valores_dia=['Lunes','Martes','Miércoles', 'Jueves','Viernes','Sábado','Domingo']
# En este orden Lunes .... Domingo aparecerán las columnas correspondientes a estas dummy variables

valores_edad=['0-15', '16-30','31-50','51_o_mas'] 
# En el mismo orden aparecerán las columnas de estas dummy variables.

tdummy5=("onehot",OneHotEncoder(categories=[valores_dia,valores_edad],sparse=False), ['dia','edad'])
```

Creamos el transformcdor de columnas igual que antes:

```python
transformador_columnas_dummy5 = ColumnTransformer(transformers=[tdummy5],remainder='passthrough')
```

**Pero esta vez lo entrenamos en el X_train**  

La siguiente es la única línea de código que cambiará:

```python
transformador_columnas_dummy5.fit(X_train5)

```

```python
X_train_dummy5=transformador_columnas_dummy5.transform(X_train5)

X_test_dummy5=transformador_columnas_dummy5.transform(X_test5)

display(X_train_dummy5)
display(X_test_dummy5)
```

Ahora que hemos visto que podemos acceder a los nombres de las variables generadas con OneHotEncoder, podemos ver qué columnas tenemos:


```python
columnas=transformador_columnas_dummy5.get_feature_names()
columnas
```

Las primeras 7 columnas, corresponden a las dummy variables derivadas de 'dia', las 4 siguientes a las dereivadas de "edad" y la que no fue afectada, x1, es la última de todas.  

Compare con el resultado obtenido anteriormente y verá que las columnas están en distinto orden. 

Si lo necesitamos podemos rearmar el dataframe:  
    

```python
dfXtrain_transf=pd.DataFrame(X_train_dummy5, columns=columnas)
dfXtrain_transf
```
